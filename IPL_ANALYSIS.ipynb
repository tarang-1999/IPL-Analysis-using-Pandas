```python
# Find number of matches played in all cities
df['city'].value_counts()
```

```python
# Find all matches played in hyd
hyd = df[df['city'] == 'Hyderabad']
hyd['city'].value_counts()['Hyderabad']
```

```python
# Find matches played in hyd after 2010
hyd = df[(df['city'] == 'Hyderabad') & (df['date'] >= '2010-01-01')]
hyd['city'].value_counts()['Hyderabad']

# or
hyd = df['city'] == 'Hyderabad'
yr = df['date'] >= '2010-01-01'
a = df[hyd & yr]
a['city'].value_counts()['Hyderabad']
```

```python
# How many matches did all teams played win
graph = df['winner'].value_counts()
graph
# value counts always works on series
```

```python
# Represent above with a horizontal bar graph
graph.plot(kind = 'barh')
# For vertical it is (bar)
```
